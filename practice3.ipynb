```python
df["Age"] = np.where(df['Age'] < 18, np.nan, df['Age'])
df
```

```python
df["Age"] = np.where(df['Age'] > 100, np.nan, df['Age'])
df
```

```python
df["PurchaseAmount"] = np.where(df['PurchaseAmount'] < 0, np.nan, df['PurchaseAmount'])
df
```

```python
mapping = {'Naija':"Nigeria","GH" : 'Ghana',"NGR" : "Nigeria"}
df['Country'] = df['Country'].replace(mapping)
df
```

```python
df_backup = df.copy()
```

```python
df.loc[~df['Email'].str.contains('@', na = False), 'Email'] = 'Invalid'
df
```

```python
df.loc[~df['Email'].str.contains('.', na = False), 'Email'] = 'Invalid'
df
```

```python
df['SignupDate'] = pd.to_datetime(df['SignupDate'], errors = 'coerce')
df
```
