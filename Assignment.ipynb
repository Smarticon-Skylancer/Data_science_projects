next to clean the "Financing Type Normalized" column we can join the "Categories" values to Fill up the missing values in the "Financing Type Normalized Column"

```python
import ast

# Convert Categories strings to lists
def parse_categories(x):
    try:
        if isinstance(x, str) and x.startswith("["):
            return ast.literal_eval(x)
        else:
            return []
    except:
        return []

df["Categories"] = df["Categories"].apply(parse_categories)

# Fill missing financing types with the first category
df["Financing Type Normalized"] = df.apply(
    lambda row: row["Categories"][0] if pd.isna(row["Financing Type Normalized"]) and len(row["Categories"]) > 0 else row["Financing Type Normalized"],
    axis=1
)
## i asked Chat gpt for the code not mine
```

```python
df['Financing Type Normalized'].fillna('none',inplace=True)
df
```

Then finally ill drop the Categories column cause i have already appended the categories to the Financing type 

```python
df.drop(columns =['Categories'], inplace = True)
df
```

Finally i would like to change the "Found at" column to date and time for better Understanding.

```python
df["Found At"]= pd.to_datetime(df['Found At'], errors = 'coerce')
df
```

In adition if the data was for a machine Learning purpose then ill change all categorical data into numerical data for the ML model to learn from it

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_data = ['Website Domain','Found At','Financing Type Normalized']
for data in categorical_data:
    df[data]= le.fit_transform(df[data])
df
```
